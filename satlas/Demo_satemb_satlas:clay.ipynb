```python
def chop_geotiff(input_path, output_dir, tile_size):
    # Open the input GeoTIFF file
    with rasterio.open(input_path) as src:
        width = src.width
        height = src.height

        # Calculate number of tiles in x and y directions
        x_tiles = width // tile_size
        y_tiles = height // tile_size

        # Iterate over the rows and columns to create tiles
        for i in range(x_tiles):
            for j in range(y_tiles):
                window = Window(i * tile_size, j * tile_size, tile_size, tile_size)
                transform = src.window_transform(window)
                tile_data = src.read(window=window)

                # Define output file name and path
                output_file = os.path.join(output_dir, f'tile_{i}_{j}.tif')

                # Save the tile to a new GeoTIFF file
                with rasterio.open(
                    output_file, 'w',
                    driver='GTiff',
                    height=tile_data.shape[1],
                    width=tile_data.shape[2],
                    count=src.count,
                    dtype=tile_data.dtype,
                    crs=src.crs,
                    transform=transform
                ) as dst:
                    dst.write(tile_data)
                    
                print(f'Saved {output_file}')
```

```python
# Define input GeoTIFF path, output directory, and tile size
input_geotiff = "/Users/bmetzler/Documents/satemb_cocoa/demo_data/ci_raster_32629.tif"
output_directory = "/Users/bmetzler/Documents/satemb_cocoa/demo_data/tiles"
tile_size = 420

# Create output directory if it doesn't exist
os.makedirs(output_directory, exist_ok=True)

# Chop the GeoTIFF into tiles
chop_geotiff(input_geotiff, output_directory, tile_size)
```
